```python
# get information about median income, education, and population per cbsa area
def get_cbsa_data(cbsa_code_lst, year_lst, api_key):
    cbsa_data_df = pd.DataFrame(columns=['Median Income', 'Education', 'Population', 'Name', 'cbsa', 'year'])
    for year in year_lst:
        count = 0
        print('-'*25, year, '-'*25)
        for cbsa_code in cbsa_code_lst:
            count += 1
            print(count)
            try:
                temp_df = pd.read_json('https://api.census.gov/data/'+year+'/acs/acs5?get=B19013_001E,B15003_022E,B01003_001E,NAME&for=metropolitan%20statistical%20area/micropolitan%20statistical%20area:'
                                       +str(int(cbsa_code))
                                       +'&key='+api_key).drop(0)
                temp_df.columns=['Median Income', 'Education', 'Population', 'Name', 'cbsa']
                temp_df['year']=year
                cbsa_data_df = cbsa_data_df.append(temp_df)
            except:
                pass
    return cbsa_data_df  
```

```python
# get information about median income, education, and population per zcta area
def get_zip_data(zcta_code_lst, year_lst, api_key):
    zcta_data_df = pd.DataFrame(columns=['Median Income', 'Education', 'Population', 'Name', 'zip', 'year'])
    for year in year_lst:
        count = 0
        print('-'*25, year, '-'*25)
        for zcta_code in zcta_code_lst:
            if len(str(zcta_code)) < 5:
                zcta_code = '0'*(5-len(str(zcta_code)))+str(zcta_code)
            count += 1
            print(count)
            try:
                temp_df = pd.read_json('https://api.census.gov/data/'+year+'/acs/acs5?get=B19013_001E,B15003_022E,B01003_001E,NAME&for=zip%20code%20tabulation%20area:'
                                       +str(zcta_code)
                                       +'&key='+api_key).drop(0)
                temp_df.columns=['Median Income', 'Education', 'Population', 'Name', 'zip']
                temp_df['year']=year
                zcta_data_df = zcta_data_df.append(temp_df)
            except:
                pass
    return zcta_data_df
```

```python
cleaned_grouped_df = cleaned_grouped_df[['year','RegionName_zip','ZHVI_MiddleTier_zip', 'ZHVI_MiddleTier_metro', 'Median Income_acs_zip',
                   'Median Income_acs_cbsa', 'Education_acs_zip', 'Education_acs_cbsa', 'cbsa', 'eligible_gentrification',
                                        'Population_acs_zip', 'Population_acs_cbsa']]
```

```python
cleaned_grouped_df['Education_pct_zip'] = cleaned_grouped_df['Education_acs_zip']/cleaned_grouped_df['Population_acs_zip']
```

```python
cleaned_grouped_df['Education_pct_cbsa'] = cleaned_grouped_df['Education_acs_cbsa']/cleaned_grouped_df['Population_acs_cbsa']
```
